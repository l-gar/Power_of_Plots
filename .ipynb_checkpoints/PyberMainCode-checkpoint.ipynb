### Observable Trends

* Although the Average fare is the lowest in urban cities it also contains the highest number of total rides per year. 

* Urban cities control 62.7 % of the market share of total Fares per city type.

* Urban cities also contain 80% of the total amount of drivers. 

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
city_data = 'Pyber/data/city_data.csv'
ride_data = 'Pyber/data/ride_data.csv'
```

```python
city_df = pd.read_csv(city_data)
ride_df = pd.read_csv(ride_data)
```

```python
all_data = pd.merge(city_df, ride_df, on = 'city')
all_data.head(5)

#all_data[all_data['city']== 'Lake Jonathanshire']['fare'].mean()
```

```python
urban_df = all_data.loc[all_data['type'] == "Urban"]
urban_df = pd.DataFrame(urban_df)
urban_city_df = urban_df.groupby('city')

urban_favg = urban_city_df['fare'].mean()
urban_favg = pd.DataFrame(urban_favg)
#urban_favg.head()

urban_rides = urban_city_df['ride_id'].count()
urban_rides = pd.DataFrame(urban_rides)
#urban_rides.head()

urban_drivers = urban_city_df['driver_count'].max()
urban_drivers = pd.DataFrame(urban_drivers)
#urban_drivers.head(50)

urban_favg_rides = pd.merge(urban_favg, urban_rides, on = 'city')
#urban_data.head(50)

urban_data = pd.merge(urban_favg_rides, urban_drivers, on = 'city')
urban_data.head(5)
```

```python
suburban_df = all_data.loc[all_data['type'] == "Suburban"]
suburban_df = pd.DataFrame(suburban_df)
suburban_city_df = suburban_df.groupby('city')

suburban_favg = suburban_city_df['fare'].mean()
suburban_favg = pd.DataFrame(suburban_favg)
#urban_favg.head()

suburban_rides = suburban_city_df['ride_id'].count()
suburban_rides = pd.DataFrame(suburban_rides)
#urban_rides.head()

suburban_drivers = suburban_city_df['driver_count'].max()
suburban_drivers = pd.DataFrame(suburban_drivers)
#urban_drivers.head(50)

suburban_favg_rides = pd.merge(suburban_favg, suburban_rides, on = 'city')
#urban_data.head(50)

suburban_data = pd.merge(suburban_favg_rides, suburban_drivers, on = 'city')
suburban_data.head(5)
```

```python
rural_df = all_data.loc[all_data['type'] == "Rural"]
rural_df = pd.DataFrame(rural_df)
rural_city_df = rural_df.groupby('city')

rural_favg = rural_city_df['fare'].mean()
rural_favg = pd.DataFrame(rural_favg)
#urban_favg.head()

rural_rides = rural_city_df['ride_id'].count()
rural_rides = pd.DataFrame(rural_rides)
#urban_rides.head()

rural_drivers = rural_city_df['driver_count'].max()
rural_drivers = pd.DataFrame(rural_drivers)
#urban_drivers.head(50)

rural_favg_rides = pd.merge(rural_favg, rural_rides, on = 'city')
#urban_data.head(50)

rural_data = pd.merge(rural_favg_rides, rural_drivers, on = 'city')
rural_data.head(5)
```

## Bubble Plot of Ride Sharing Data

```python
x = urban_data['ride_id']
y = urban_data['fare']
x2 = suburban_data['ride_id']
y2 = suburban_data['fare']
x3 = rural_data['ride_id']
y3 = rural_data['fare']

plt.scatter(x, y, s= urban_data['driver_count'] * 10, c = 'gold' , label = 'Urban', edgecolor = 'black', alpha=.6, linewidths=.4)
plt.scatter(x2, y2, s= suburban_data['driver_count'] * 10, c = 'lightskyblue', label = 'Suburban', edgecolor='black', alpha=.6, linewidths=.4)
plt.scatter(x3, y3, s= rural_data['driver_count'] * 10, c = 'lightcoral', label = 'Rural', edgecolor = 'black', alpha= .6, linewidths=.4 )

plt.title('Pyber ride Sharing Data (2016)')
plt.xlabel('Total Number of Rides')
plt.ylabel('Average Fare ($)')
plt.legend(prop={'size': 10}, markerscale = .75)
plt.figtext( 1,.55,('Note: \nCircle size correlates with driver count per city.'))
plt.grid(bool)
plt.tight_layout()

plt.savefig("Pyber_RideSharing_fig")

plt.show()
```

## Total Fares by City Type

```python
urban_tf = urban_df['fare'].sum()
suburban_tf = suburban_df['fare'].sum()
rural_tf = rural_df['fare'].sum()

pie1_names = ['Urban', 'Suburban','Rural']
pie1_variables = [urban_tf, suburban_tf, rural_tf]
colors = ['gold','lightskyblue', 'lightcoral']
explode = [0.1,0,0]

plt.pie(pie1_variables, labels = pie1_names, colors = colors, explode = explode, shadow= True, autopct='%.1f%%', pctdistance=.5, labeldistance=1.1, startangle=-70)

plt.title('% of Total Fares by City Type')
plt.axis('auto')
plt.savefig("Percent_total_Fares_fig")

plt.show
```

## Total Rides by City Type

```python
# total rides count
urban_tr = urban_df['ride_id'].count()
suburban_tr = suburban_df['ride_id'].count()
rural_tr = rural_df['ride_id'].count()

pie2_names = ['Urban', 'Suburban','Rural']
pie2_variables = [urban_tr, suburban_tr, rural_tr]

plt.pie(pie2_variables, labels = pie2_names, colors = colors, explode = explode, shadow= True, autopct='%.1f%%', pctdistance=.5, labeldistance=1.1, startangle=-70)

plt.title('% of Total Rides by City Type')
plt.axis('auto')
plt.savefig("Percent_total_Rides_fig")

plt.show
```
